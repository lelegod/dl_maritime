# Data preprocessing

```python
# Configuration Parameters
GAP_BREAK_MIN = 180
INTERPOLATION_LIMIT_MIN = None
MAX_DISTANCE_M = 3000
MAX_SOG_KNOTS = 40
OUTPUT_PATH = "data/ais_data_5min_clean.csv"
NUM_COLS = ["SOG", "COG", "Longtitude", "Latitude"]
MIN_SEGMENT_LENGTH = 25
INTERVAL = 5
```

```python
print("="*60)
print("STEP 1: Data Preprocessing")
print("="*60)

# Sort data by MMSI and Timestamp
df_cargo = df_cargo.sort_values(["MMSI", "Timestamp"]).reset_index(drop=True)
df_cargo["Timestamp"] = pd.to_datetime(df_cargo["Timestamp"], errors="coerce")

print(f"Initial data shape: {df_cargo.shape}")
print(f"Data types:\n{df_cargo.dtypes}\n")

# Segment trajectories based on time gaps
print("Segmenting trajectories...")
df = segment_and_renumber(df_cargo, GAP_BREAK_MIN)

# Downsample & interpolate per segment
print(f"Downsampling to {INTERVAL}-minute intervals and interpolating...")
results = []

for (mmsi, seg), g in df.groupby(["MMSI", "Segment"], observed=True):
    g = g.set_index("Timestamp")
    
    # Downsample to 5-minute intervals (keep last observation)
    g1 = g.resample(f"{INTERVAL}min").last()
    
    # Interpolate numeric columns for short gaps only
    g1[NUM_COLS] = g1[NUM_COLS].interpolate(
        method="time", limit=INTERPOLATION_LIMIT_MIN, limit_direction="both"
    )
    
    # Drop rows where ANY of the critical columns are NaN
    g1 = g1.dropna(subset=NUM_COLS, how="any")
    
    # Fill identifiers
    g1["MMSI"] = mmsi
    g1["Segment"] = seg
    
    # Calculate distance and speed between consecutive points
    lat = g1["Latitude"].to_numpy()
    lon = g1["Longtitude"].to_numpy()
    lat_prev, lon_prev = np.roll(lat, 1), np.roll(lon, 1)
    lat_prev[0], lon_prev[0] = lat[0], lon[0]
    
    g1["distance_m"] = haversine_m(lat, lon, lat_prev, lon_prev)
    g1.loc[g1.index[0], "distance_m"] = 0.0
    g1["speed_mps_track"] = g1["distance_m"] / 60.0
    
    # Filter unrealistic movement or SOG
    g1 = g1[(g1["distance_m"] < MAX_DISTANCE_M) & (g1["SOG"] <= MAX_SOG_KNOTS)]
    
    results.append(g1)

# Combine all segments
df_clean = pd.concat(results).reset_index()
```

```python
# Compute dSOG and dCOG
def compute_delta_kinematics(df):
    """Add dSOG and dCOG features per segment."""
    dfs = []

    for (mmsi, seg), g in df.groupby(["MMSI", "Segment"], observed=True):
        g = g.sort_values("Timestamp").copy()

        # Difference (delta) in SOG between the current and previous time step
        g["dSOG"] = g["SOG"].diff().fillna(0)

        # Handle 360-degree wrap-around for COG difference
        # COG is in degrees
        dCOG_raw = g["COG"].diff()
        dCOG_wrapped = (dCOG_raw + 540) % 360 - 180
        g["dCOG"] = dCOG_wrapped.fillna(0)

        dfs.append(g)

    return pd.concat(dfs, ignore_index=True)
def haversine(lat1, lon1, lat2, lon2):
    '''
    Calculates the great-circle distance between two points (lat1, lon1)
    and (lat2, lon2) using the Haversine formula (in meters)
    '''
    R = 6371000.0  # Earth radius in meters
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat / 2) ** 2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2) ** 2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Mean Absolute Error between the predicted speed and the true speed of a vessel
def compute_speed_error(pred_coords, true_coords, seconds=300.0):
    '''
    pred_coords/true_coords shape: (N, 2) -> (Latitude, Longtitude)
    Calculates the speed error across the predicted sequence (in m/s).
    '''
    speeds_pred = []
    speeds_true = []

    # Need at least 2 points to calculate speed (N - 1 intervals)
    n = min(len(pred_coords) - 1, len(true_coords) - 1)

    for i in range(n):
        dp = haversine(pred_coords[i][0], pred_coords[i][1],
                       pred_coords[i + 1][0], pred_coords[i + 1][1])
        dt = haversine(true_coords[i][0], true_coords[i][1],
                       true_coords[i + 1][0], true_coords[i + 1][1])

        speeds_pred.append(dp / seconds)
        speeds_true.append(dt / seconds)

    if not speeds_pred:
        return 0.0

    return np.mean(np.abs(np.array(speeds_pred) - np.array(speeds_true)))

# Quantifies the prediction error in how sharply the vessel turns
def compute_turn_rate_error(pred_coords, true_coords):
    '''
    pred_coords/true_coords shape: (N, 2) -> (Latitude, Longtitude)
    Calculates the mean absolute turn rate error (in degrees).
    '''
    def compute_cog_from_coords(lat, lon):
        dlat = np.diff(lat)
        dlon = np.diff(lon)
        # arctan2 returns results in radians; convert to degrees
        return np.degrees(np.arctan2(dlon, dlat))

    n = min(len(pred_coords), len(true_coords))
    pred = pred_coords[:n]
    true = true_coords[:n]

    # Calculate COG for all steps (N-1 values)
    pred_cog = compute_cog_from_coords(pred[:, 0], pred[:, 1])
    true_cog = compute_cog_from_coords(true[:, 0], true[:, 1])

    if len(pred_cog) < 2 or len(true_cog) < 2:
        return 0.0

    # Turn rate is the difference between successive COG values (N-2 values)
    pred_turn = np.diff(pred_cog)
    true_turn = np.diff(true_cog)

    n2 = min(len(pred_turn), len(true_turn))
    # Calculates the mean absolute difference in turn rate (in degrees)
    return np.mean(np.abs(pred_turn[:n2] - true_turn[:n2]))
```

# Transformer model

```python
'''
Define Transformer Encoder
Attention mechanism allows the model to learn that predicting the next turn might rely heavily on the last few steps of dCOG (change in course) and COG (current course),
while predicting speed might rely more on the current SOG (speed over ground) and the dSOG history
'''
class AISTransformer(nn.Module):
    '''
    Embedding size -> embed_Size
    Attention heads allowing the model to focus on different parts of the input window simultaneously -> num_heads
    Feed-Forward network Dimension (2 layers MLP) -> ff_dim
    '''
    def __init__(self, n_features, n_outputs, embed_size=64, num_heads=2,
                 num_layers=2, ff_dim=128, dropout=0.1):
        super().__init__()

        self.embedding = nn.Linear(n_features, embed_size)

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=embed_size,
            nhead=num_heads,
            dim_feedforward=ff_dim,
            dropout=dropout,
            batch_first=True
        )
        self.encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.encoder.gradient_checkpointing = True

        # predict 1-step delta at a time (autoregressive decoder)
        self.decoder = nn.Linear(embed_size, n_outputs)


    def forward(self, x):
        # The input x is (B,W,F) and should be float32, which is ensured by the fix.
        z = self.embedding(x)
        z = self.encoder(z)
        # Apply the decoder to the last time-step representation
        last_timestep_output = z[:, -1, :]
        prediction = self.decoder(last_timestep_output) # (B, 64) -> (B, 4)
        return prediction


# Build the model
print("="*60)
print("Building PyTorch Transformer Model")
print("="*60)

model = AISTransformer(
    n_features=len(FEATURES),
    n_outputs=len(TARGET_FEATURES)
).to(DEVICE)

# Display model architecture
print(model)
print(f"\nModel moved to: {DEVICE}")

# Count parameters
total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"Total parameters: {total_params:,}")
print(f"Trainable parameters: {trainable_params:,}")
```

```python
'''
Define Transformer Encoder
Attention mechanism allows the model to learn that predicting the next turn might rely heavily on the last few steps of dCOG (change in course) and COG (current course),
while predicting speed might rely more on the current SOG (speed over ground) and the dSOG history
'''
class AISTransformer(nn.Module):
    '''
    Embedding size -> embed_Size
    Attention heads allowing the model to focus on different parts of the input window simultaneously -> num_heads
    Feed-Forward network Dimension (2 layers MLP) -> ff_dim
    '''
    def __init__(self, input_size, output_size, embed_size=64, num_heads=2,
                 num_layers=2, ff_dim=128, dropout=0.1):
        super().__init__()

        self.embedding = nn.Linear(input_size, embed_size)

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=embed_size,
            nhead=num_heads,
            dim_feedforward=ff_dim,
            dropout=dropout,
            batch_first=True
        )
        self.encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.encoder.gradient_checkpointing = True

        # predict 1-step delta at a time (autoregressive decoder)
        self.decoder = nn.Linear(embed_size, output_size)

    def forward(self, x):
        """Forward only encodes input window; decoding happens outside."""
        # The input x is (B,W,F) and should be float32, which is ensured by the fix.
        z = self.embedding(x)
        z = self.encoder(z)
        return z[:, -1, :]  # last timestep representation
```

```python
# Setup training components
print("="*60)
print("Setting up Training Components")
print("="*60)

# Create output directory - use different folder for delta model
OUTPUT_DIR = "models/Transformer_delta"
os.makedirs(OUTPUT_DIR, exist_ok=True)
os.makedirs(f"{OUTPUT_DIR}/plots", exist_ok=True)

print(f"Output directory: {OUTPUT_DIR}")

# Loss function (Huber Loss / Smooth L1)
criterion = nn.SmoothL1Loss()

# Optimizer
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# Learning rate scheduler (similar to ReduceLROnPlateau)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode='min', factor=0.5, patience=5, min_lr=1e-6
)

print("Training components configured:")
print("  - Loss: Smooth L1 (Huber)")
print("  - Optimizer: Adam (lr=0.001)")
print("  - Scheduler: ReduceLROnPlateau (factor=0.5, patience=5)")
print("  - Early Stopping: patience=30")
```

```python
# Training loop
print("="*60)
print("Training PyTorch Transformer Model")
print("="*60)

EPOCHS = 1000
PATIENCE = 15

# Training history
history = {
    'train_loss': [],
    'val_loss': [],
    'train_mae': [],
    'val_mae': [],
    'lr': []
}

best_val_loss = float('inf')
patience_counter = 0
best_model_state = None

for epoch in range(EPOCHS):
    # Training phase
    model.train()
    train_loss = 0.0
    train_mae = 0.0
    
    for batch_X, batch_y in train_loader:
        optimizer.zero_grad()
        outputs = model(batch_X)
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()
        
        train_loss += loss.item()
        train_mae += torch.mean(torch.abs(outputs - batch_y)).item()
    
    train_loss /= len(train_loader)
    train_mae /= len(train_loader)
    
    # Validation phase
    model.eval()
    val_loss = 0.0
    val_mae = 0.0
    
    with torch.no_grad():
        for batch_X, batch_y in val_loader:
            outputs = model(batch_X)
            loss = criterion(outputs, batch_y)
            val_loss += loss.item()
            val_mae += torch.mean(torch.abs(outputs - batch_y)).item()
    
    val_loss /= len(val_loader)
    val_mae /= len(val_loader)
    
    # Update scheduler
    scheduler.step(val_loss)
    current_lr = optimizer.param_groups[0]['lr']
    
    # Save history
    history['train_loss'].append(train_loss)
    history['val_loss'].append(val_loss)
    history['train_mae'].append(train_mae)
    history['val_mae'].append(val_mae)
    history['lr'].append(current_lr)
    
    # Early stopping check
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        patience_counter = 0
        best_model_state = model.state_dict().copy()
        # Save best model
        torch.save(best_model_state, f'{OUTPUT_DIR}/best_transformer_model.pth')
        print(f"Epoch {epoch+1}/{EPOCHS} - loss: {train_loss:.4f} - mae: {train_mae:.4f} - val_loss: {val_loss:.4f} - val_mae: {val_mae:.4f} - lr: {current_lr:.6f} ✓ saved")
    else:
        patience_counter += 1
        print(f"Epoch {epoch+1}/{EPOCHS} - loss: {train_loss:.4f} - mae: {train_mae:.4f} - val_loss: {val_loss:.4f} - val_mae: {val_mae:.4f} - lr: {current_lr:.6f}")
    
    if patience_counter >= PATIENCE:
        print(f"\nEarly stopping triggered after {epoch+1} epochs")
        break

# Restore best model
if best_model_state is not None:
    model.load_state_dict(best_model_state)
    print(f"\nRestored best model with val_loss: {best_val_loss:.4f}")

# Save final model and history
torch.save(model.state_dict(), f'{OUTPUT_DIR}/final_transformer_model.pth')
with open(f'{OUTPUT_DIR}/training_history.json', 'w') as f:
    json.dump(history, f)

print("\n" + "="*60)
print("Training Complete!")
print("="*60)
```

```python
# Analyze Error Accumulation Across Prediction Steps (Delta Model) - FIXED
print("="*60)
print("Analyzing Error Accumulation Across Prediction Horizons (Delta Model)")
print("="*60)

from collections import defaultdict

# Build mapping of test sequences to their ground truth future values
segment_to_indices = defaultdict(list)
for idx, orig_idx in enumerate(test_indices):
    seg_info = segment_info[orig_idx]
    key = (seg_info['mmsi'], seg_info['segment'])
    segment_to_indices[key].append((idx, orig_idx))

# For sequences with consecutive ground truth, compute multi-step errors
step_errors = {step: [] for step in range(PREDICTION_STEPS)}
sequences_with_gt = 0

for key, indices in segment_to_indices.items():
    # Sort by original index to ensure temporal order
    indices_sorted = sorted(indices, key=lambda x: x[1])
    
    for i, (test_idx, orig_idx) in enumerate(indices_sorted):
        # For step N prediction, we need ground truth N steps ahead
        # Step 1: compare with current sequence's target (t+1)
        # Step 2: compare with next sequence's target (t+2)
        # etc.
        
        # Step 0 (first prediction) compares with current sequence's target
        preds = all_iterative_preds_abs[test_idx]  # (n_steps, 4)
        
        # Step 0: compare with THIS sequence's target
        gt_lat_step0 = original_positions_test.iloc[test_idx]['target_lat']
        gt_lon_step0 = original_positions_test.iloc[test_idx]['target_lon']
        dist_error_step0 = haversine_m(preds[0, 0], preds[0, 1], gt_lat_step0, gt_lon_step0)
        step_errors[0].append(dist_error_step0)
        
        # For steps 1+, we need future sequences' targets
        available_future = len(indices_sorted) - i - 1
        
        if available_future > 0:
            sequences_with_gt += 1
            
            # Step N (N>=1): compare with sequence[i+N]'s target
            for step in range(1, min(PREDICTION_STEPS, available_future + 1)):
                future_test_idx = indices_sorted[i + step][0]
                
                # Ground truth absolute position
                gt_lat = original_positions_test.iloc[future_test_idx]['target_lat']
                gt_lon = original_positions_test.iloc[future_test_idx]['target_lon']
                
                # Predicted absolute position
                pred_lat = preds[step, 0]
                pred_lon = preds[step, 1]
                
                # Calculate distance error
                dist_error = haversine_m(pred_lat, pred_lon, gt_lat, gt_lon)
                step_errors[step].append(dist_error)

print(f"Sequences with at least 1 future ground truth: {sequences_with_gt}")
print(f"\n{'Step':<6} {'Time':>10} {'Samples':>10} {'Mean (m)':>12} {'Median (m)':>12} {'Std (m)':>12}")
print("-" * 70)

step_stats = []
for step in range(PREDICTION_STEPS):
    errors = step_errors[step]
    if len(errors) > 0:
        mean_err = np.mean(errors)
        median_err = np.median(errors)
        std_err = np.std(errors)
        time_ahead = (step + 1) * INTERVAL
        step_stats.append({
            'step': step + 1,
            'time_min': time_ahead,
            'n_samples': len(errors),
            'mean_m': mean_err,
            'median_m': median_err,
            'std_m': std_err,
            'p90_m': np.percentile(errors, 90),
            'p95_m': np.percentile(errors, 95)
        })
        print(f"{step+1:<6} {time_ahead:>7} min {len(errors):>10} {mean_err:>12.2f} {median_err:>12.2f} {std_err:>12.2f}")

# Plot error accumulation
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

times = [s['time_min'] for s in step_stats]
means = [s['mean_m'] / 1000 for s in step_stats]  # Convert to km
medians = [s['median_m'] / 1000 for s in step_stats]
stds = [s['std_m'] / 1000 for s in step_stats]

# Plot 1: Mean and Median error over time
axes[0].plot(times, means, 'b-o', label='Mean Error', linewidth=2, markersize=8)
axes[0].plot(times, medians, 'g-s', label='Median Error', linewidth=2, markersize=8)
axes[0].fill_between(times, 
                      [m - s for m, s in zip(means, stds)],
                      [m + s for m, s in zip(means, stds)],
                      alpha=0.2, color='blue', label='±1 Std Dev')
axes[0].set_xlabel('Prediction Horizon (minutes)', fontsize=12)
axes[0].set_ylabel('Distance Error (km)', fontsize=12)
axes[0].set_title('Error Accumulation - Delta Coordinate Model', fontsize=14)
axes[0].legend()
axes[0].grid(True, alpha=0.3)
axes[0].set_xticks(times)

# Plot 2: Error distribution at different horizons
selected_steps = [0, 4, 9]  # 5min, 25min, 50min
colors = ['green', 'orange', 'red']
for step, color in zip(selected_steps, colors):
    if step < len(step_errors) and len(step_errors[step]) > 0:
        errors_km = np.array(step_errors[step]) / 1000
        axes[1].hist(errors_km, bins=50, alpha=0.5, label=f'{(step+1)*INTERVAL} min', color=color, density=True)

axes[1].set_xlabel('Distance Error (km)', fontsize=12)
axes[1].set_ylabel('Density', fontsize=12)
axes[1].set_title('Error Distribution at Different Horizons', fontsize=14)
axes[1].legend()
axes[1].grid(True, alpha=0.3)
axes[1].set_xlim(0, 10)  # Adjusted scale

plt.tight_layout()
plt.savefig(f'{OUTPUT_DIR}/plots/multi_step_error_analysis.png', dpi=150)
plt.show()

# Summary statistics
print(f"\n{'='*60}")
print("Multi-Step Prediction Summary (Delta Model)")
print("="*60)
print(f"Prediction horizon: {PREDICTION_STEPS * INTERVAL} minutes ({PREDICTION_STEPS} steps)")
print(f"\nAt 5 min:  Mean = {step_stats[0]['mean_m']:.0f} m, Median = {step_stats[0]['median_m']:.0f} m")
if len(step_stats) >= 5:
    print(f"At 25 min: Mean = {step_stats[4]['mean_m']:.0f} m, Median = {step_stats[4]['median_m']:.0f} m")
if len(step_stats) >= 10:
    print(f"At 50 min: Mean = {step_stats[9]['mean_m']:.0f} m, Median = {step_stats[9]['median_m']:.0f} m")

# Save multi-step metrics
multi_step_metrics = {
    'model_type': 'delta_coordinates',
    'prediction_steps': PREDICTION_STEPS,
    'interval_minutes': INTERVAL,
    'total_horizon_minutes': PREDICTION_STEPS * INTERVAL,
    'step_statistics': step_stats
}

with open(f'{OUTPUT_DIR}/multi_step_metrics.json', 'w') as f:
    json.dump(multi_step_metrics, f, indent=2)

print(f"\nPlot saved to: {OUTPUT_DIR}/plots/multi_step_error_analysis.png")
print(f"Metrics saved to: {OUTPUT_DIR}/multi_step_metrics.json")
```

```python
from collections import defaultdict
from tqdm import tqdm

# Interval in seconds (5 minutes)
SECONDS_PER_INTERVAL = INTERVAL * 60
PREDICTION_STEPS = 10 # Defined in Cell 28

# Rebuild segment-to-index mapping (from multi-step analysis setup)
segment_to_indices = defaultdict(list)
for idx, orig_idx in enumerate(test_indices):
    seg_info = segment_info[orig_idx]
    key = (seg_info['mmsi'], seg_info['segment'])
    segment_to_indices[key].append((idx, orig_idx))

all_speed_errors = []
all_turn_errors = []
all_mean_haversine_errors = []

all_speed_errors = []
all_turn_errors = []

SECONDS_PER_INTERVAL = INTERVAL * 60

for key, indices in tqdm(segment_to_indices.items(), desc="Calculating Dynamic Metrics"):
    indices_sorted = sorted(indices, key=lambda x: x[1])

    for i, (test_idx, orig_idx) in enumerate(indices_sorted):
        available_future = len(indices_sorted) - i - 1
        gt_length = min(PREDICTION_STEPS, available_future + 1)

        if gt_length >= 1:
            preds = all_iterative_preds_abs[test_idx]

            # Last Input Point (Start of sequence, retrieved as a (1, 2) array)
            gt_start_point = original_positions_test.iloc[test_idx][['last_lat', 'last_lon']].to_numpy().reshape(1, 2).astype(np.float64)

            # Gather the next (gt_length) absolute target points
            gt_future_coords_list = []
            for step in range(1, gt_length + 1):
                future_test_idx = indices_sorted[i + step - 1][0]

                # Safely extract (target_lat, target_lon) as a (1, 2) array
                future_coords = original_positions_test.iloc[future_test_idx][['target_lat', 'target_lon']].to_numpy().reshape(1, 2).astype(np.float64)
                gt_future_coords_list.append(future_coords)

            # Stack the future coordinates into one (N, 2) array
            gt_future_coords_array = np.vstack(gt_future_coords_list)

            # Vstack the start point (1, 2) with the future points (N, 2)
            gt_sequence = np.vstack([gt_start_point, gt_future_coords_array])

            # Stack the start point (1, 2) with the predicted future coordinates (N, 2)
            pred_sequence = np.vstack([gt_start_point, preds[:gt_length, :2]])

            if len(gt_sequence) >= 2:
                # Speed Error (in m/s)
                speed_err = compute_speed_error(pred_sequence, gt_sequence, seconds=SECONDS_PER_INTERVAL)
                all_speed_errors.append(speed_err)

                # Turn-Rate Error (deg)
                turn_err = compute_turn_rate_error(pred_sequence, gt_sequence)
                all_turn_errors.append(turn_err)

# Calculate the mean value across ALL test sequences
mean_speed_err = np.mean(all_speed_errors)
mean_turn_err = np.mean(all_turn_errors)
mean_positional_error = np.mean([s['mean_m'] for s in step_stats])


print("\n" + "="*60)
print("Multi-Step (50 min) Dynamic Performance Summary")
print("="*60)
print(f"Sequences evaluated (>= 2 steps): {len(all_speed_errors)}\n")

print(f"{'Metric':<30} {'Mean Value':>15}")
print("-" * 46)
print(f"{'Avg Positional Error (m)':<30} {mean_positional_error:>15.2f}")
print(f"{'Mean Speed Error (m/s)':<30} {mean_speed_err:>15.4f}")
print(f"{'Mean Turn-Rate Error (deg)':<30} {mean_turn_err:>15.4f}")
print("=" * 46)
```
